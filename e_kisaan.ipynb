```python
import customtkinter as ctk
import speech_recognition as sr
from gtts import gTTS
import pygame
import threading
import tempfile
import os
import joblib
import pandas as pd
import random

# Load model
model = joblib.load("e-kisaan/model.pkl")

# Init audio
pygame.mixer.init()

# Speak using GTTS and pygame
def speak(text, lang='en'):
    tts = gTTS(text=text, lang=lang)
    with tempfile.NamedTemporaryFile(delete=False, suffix=".mp3") as fp:
        temp_path = fp.name
    try:
        tts.save(temp_path)
        pygame.mixer.music.load(temp_path)
        pygame.mixer.music.play()
        while pygame.mixer.music.get_busy():
            continue
    finally:
        os.remove(temp_path)

# Recognize speech
def listen():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        try:
            audio = r.listen(source, timeout=5)
            text = r.recognize_google(audio)
            return text.lower()
        except:
            return ""

# Process commands
def process_command(command):
    if "crop" in command:
        # Simulate sensor input (replace with Arduino later)
        ph = random.uniform(5.5, 8.0)
        moisture = random.uniform(10, 80)
        temp = random.uniform(20, 40)
        data = pd.DataFrame([[ph, moisture, temp]], columns=['Soil_pH', 'Soil_Moisture', 'Temperature_C'])
        crop = model.predict(data)[0]
        return f"The suitable crop for this soil is {crop}."

    elif "weather" in command or "mausam" in command:
        return "Current temperature is around 32¬∞C with clear skies."  # Mock response

    elif "namaste" in command or "hello" in command:
        return "Namaste! Main e-Kisaan hoon, aapki farming assistant."

    elif "kaise ho" in command:
        return "Main badhiya hoon! Aap kaise ho?"

    else:
        return "Sorry, mujhe yeh samajh nahi aaya."

# Handle voice command
def handle_voice():
    command = listen()
    if command:
        response = process_command(command)
        chatbox.insert("end", "üë©‚Äçüåæ You: " + command)
        chatbox.insert("end", "ü§ñ e-Kisaan: " + response)
        lang = 'hi' if any(word in command for word in ['namaste', 'kaise', 'mausam']) else 'en'
        speak(response, lang=lang)

# GUI setup
ctk.set_appearance_mode("light")
ctk.set_default_color_theme("green")

app = ctk.CTk()
app.geometry("600x500")
app.title("e-Kisaan - Smart Farming Assistant")

title = ctk.CTkLabel(app, text="üåø e-Kisaan", font=("Arial Black", 30))
title.pack(pady=10)

subtitle = ctk.CTkLabel(app, text="Your Smart Farming Companion (Bilingual)", font=("Arial", 16))
subtitle.pack(pady=5)

chatbox = ctk.CTkTextbox(app, width=550, height=300, font=("Arial", 14))
chatbox.pack(pady=10)

listen_btn = ctk.CTkButton(app, text="üéôÔ∏è Ask e-Kisaan", font=("Arial", 16), command=lambda: threading.Thread(target=handle_voice).start())
listen_btn.pack(pady=10)

app.mainloop()
```
